### define model

```python
class MySimpleNN():

    def __init__(self,x):
        self.input=x
        self.weights=torch.rand(x.shape[1],1,dtype=torch.float64,requires_grad=True)
        self.bias=torch.zeros(1,dtype=torch.float64,requires_grad=True)

    def forward(self,x):
        z=torch.matmul(x,self.weights)+self.bias
        y_pred=torch.sigmoid(z)
        return y_pred

    def loss_fuction(self,y_pred,y_true):
        loss= -(y_train_tensor*torch.log(y_pred)+(1-y_train_tensor)*torch.log(1-y_pred)).mean()
        return loss
    

        
```

```python
learning_rate=0.1
epochs=25
```

```python
# create a model
model=MySimpleNN(x_train_tensor)
# define loop
for i in  range(epochs):
    # forwad pass
    y_pred=model.forward(x_train_tensor)

    # calculate loss
    loss=model.loss_fuction(y_pred,y_train_tensor)

    # backward pass
    loss.backward()

    # update weights and bias
    # no_grad() is used to stop tracking the gradients
    with torch.no_grad():
        model.weights.data-=learning_rate*model.weights.grad
        model.bias.data-=learning_rate*model.bias.grad

        # zero the gradients
    model.weights.grad.zero_()
    model.bias.grad.zero_()

    # print loss in each epoch
    print(f'Epoch {i+1} loss: {loss.item()}')
```

###  evaluation

```python
with torch.no_grad():
    y_pred=model.forward(x_test_tensor)
    y_pred=(y_pred>0.9).float()
    accuracy=(y_pred==y_test_tensor).float().mean()
    print(f'Accuracy: {accuracy.item()}')
```
